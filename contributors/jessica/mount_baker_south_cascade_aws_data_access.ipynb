## AWS DEM data access

Data access to historical and modern DEMs produced over Mount Baker and South Cascade Glacier in the northern Cascade Range. All DEMs are coregistered to a modern reference DEM, projected on to a common grid, and stacked using the Geospatial Time Series Analaysis ([GTSA](https://github.com/friedrichknuth/gtsa/tree/main)) library.

**Mount Baker** (1947 - 2020)  
**South Cascade Glacier** (1958 - 2021)

The reference DEM for Mount Baker is dated 2015-08-27 and is included in the stack. This DEM is derived from WADNR lidar and provides full spatial coverage. The reference DEM for South Cascade Glacier is a composite 2013-2015 DEM and is not included in the stack. This DEM was used for coregistration purposes only as it provides greater coverage over stable ground.

```python
baker_url = 's3://petrichor/geosmart/baker.zarr/'
scg_url = 's3://petrichor/geosmart/scg.zarr/'
```

```python
baker_ds = xr.open_dataset(baker_url, chunks='auto', engine='zarr', storage_options={"anon": True})
scg_ds = xr.open_dataset(scg_url, chunks='auto', engine='zarr', storage_options={"anon": True})
```

## Applying Glacier Outlines

```python
baker_glac_outline = 'https://petrichor.s3.us-west-2.amazonaws.com/geosmart/baker_glaciers.geojson'
scg_glac_outline = 'https://petrichor.s3.us-west-2.amazonaws.com/geosmart/scg_glacier.geojson'
```

Steps:
- load the outlines --> this step is required because even though geocube can accept the geojson file as an input, you cannot supply both `output_crs` and `like` kwargs to `make_geocube`, which would be needed to properly reproject the geojson.
- check that the CRSs match
- rasterize the outlines to the geospatial grid (including reprojection)
- add the mask to the dataset

```python
baker_outlines = gpd.read_file(baker_glac_outline)
baker_outlines["mask"] = 1 # if you wanted to, you could use this to create keys for each glacier
```

```python
baker_outlines.explore()
```

```python
baker_ds.rio.crs == baker_outlines.crs
```

```python
baker_outlines_ds = make_geocube(baker_outlines.to_crs(baker_ds.crs),
                                 measurements=['mask'],
                                 like=baker_ds)
```

```python
baker_ds = baker_ds.assign(baker_outlines_ds.drop('spatial_ref'))
baker_ds
```
