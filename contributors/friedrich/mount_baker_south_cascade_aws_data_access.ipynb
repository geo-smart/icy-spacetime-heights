## AWS DEM data access

Data access to historical and modern DEMs produced over Mount Baker and South Cascade Glacier in the northern Cascade Range. All DEMs are coregistered to a modern reference DEM, projected on to a common grid, and stacked using the Geospatial Time Series Analaysis ([GTSA](https://github.com/friedrichknuth/gtsa/tree/main)) library.

**Mount Baker** (1947 - 2020)  
**South Cascade Glacier** (1958 - 2021)

The reference DEM for Mount Baker is dated 2015-08-27 and is included in the stack. This DEM is derived from WADNR lidar and provides full spatial coverage. The reference DEM for South Cascade Glacier is a composite 2013-2015 DEM and is not included in the stack. This DEM was used for coregistration purposes only as it provides greater coverage over stable ground.

### Fit GP and plot

```python
import numpy as np
import pandas as pd

import gtsa


from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import (
    RBF,
    ConstantKernel,
    ExpSineSquared,
    PairwiseKernel,
    RationalQuadratic,
    WhiteKernel,
    Matern,
)
```

### Convert datetime to decimal years

```python
dec_year_dates = [np.array([gtsa.utils.date_time_to_decyear(i) for i in j]) \
                  for j in time_stamps]
```

### Create time series to predict on

```python
prediction_time_series = gtsa.temporal.create_prediction_timeseries(start_date = '1957-09-01',
                                               end_date = '2021-09-01',
                                               dt ='M')
```

### Define a kernel

```python
k1 = 30.0 * Matern(length_scale=10.0, nu=1.5)
k2 = ConstantKernel(30) * ExpSineSquared(length_scale=1, periodicity=30)
k3 = ConstantKernel(30) * ExpSineSquared(length_scale=1, periodicity=1)

kernel = k1+k2+k3
```

```python
predictions = []
std_predictions = []

for i,v in enumerate(dec_year_dates):
    time_array = dec_year_dates[i]
    data_array = time_series_data[i]

    gaussian_process_model = gtsa.temporal.GPR_model(time_array, 
                                                     data_array, 
                                                     kernel, 
                                                     alpha=2,
                                                    )

    mean_prediction, std_prediction = gtsa.temporal.GPR_predict(gaussian_process_model, 
                                                               prediction_time_series)
    
    predictions.append(mean_prediction)
    std_predictions.append(std_prediction)
    
GPR_df = pd.DataFrame(predictions).T
GPR_df.index = prediction_time_series

GPR_std_df = pd.DataFrame(std_predictions).T
GPR_std_df.index = prediction_time_series
```

### Convert decimal years back to datetime

```python
GPR_df.index = [gtsa.utils.decyear_to_date_time(i) for i in GPR_df.index]
GPR_std_df.index = [gtsa.utils.decyear_to_date_time(i) for i in GPR_std_df.index]
```

```python
tools.plot_time_series_gallery(time_stamps,
                         time_series_data,
                                       predictions_df_list=[GPR_df],
                                       std_df_list=[GPR_std_df],
                         ylim=(-50,100),
                         sharey=True,
                         cmap='cool',
                        )
```
